```python
#training and testing data
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
import io #input and output
import os
import requests
import numpy as np
from sklearn import metrics
```

```python
df = pd.read_csv("auto-mpg.csv", na_values = ["NA","?"])
df.head()
```

```python
df.tail()
```

```python
df.isna().sum() #check whether the dataset can null values or not
```

```python
#fill na values with median valueus
```

```python
median = df["horsepower"].median() #handle the missing values
```

```python
df["horsepower"]=  df["horsepower"].fillna(median)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
#input: x
#output: y

x =  df.iloc[:, 1:8]
x.head()
```

```python
y = df.iloc[:, :1]
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 12)
```

```python
#model finally
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Activation 
from tensorflow.keras.callbacks import EarlyStopping # 

#case: if my model is not reducing error (constant), stop training the model
```

```python
x.shape
```

```python
model = Sequential()

model.add(Dense(25, input_dim = x.shape[1], activation = "relu"))
model.add(Dense(10, activation = "relu"))
model.add(Dense(1))


model.compile(loss = "mean_squared_error", optimizer = "adam")



monitor = EarlyStopping(monitor = "val_loss", min_delta = 1e-3, patience = 5,
                       verbose = 1, mode = "auto", restore_best_weights = True)

model.fit(x_train, y_train, validation_data = (x_test, y_test),
         callbacks = [monitor],
         verbose = 1,
         epochs = 1000)


#train our model
```

```python
y_test.head()
```

```python
prediction = model.predict(x_test)
```

```python
MSE = metrics.mean_squared_error(y_test, prediction)
```

```python
score = np.sqrt(MSE)
```

```python
print(score)
```
